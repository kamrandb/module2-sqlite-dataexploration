```python
import pandas as pd
import sqlite3

# Load the dataset
file_url = 'https://data.cdc.gov/api/views/2m93-xvra/rows.csv?accessType=DOWNLOAD'
healthcare_data = pd.read_csv(file_url)

# Create a connection to a new SQLite database file
conn = sqlite3.connect('healthcare.db')

# Load the dataset into the SQLite database
table_name = 'health_conditions_under_18'
healthcare_data.to_sql(table_name, conn, if_exists='replace', index=False)

# Verify the data was inserted correctly by displaying the first few rows
pd.read_sql_query("SELECT * FROM health_conditions_under_18 LIMIT 10", conn)
```

```python
# Select all records where the condition is 'Current asthma among persons under 18 years'
query1 = "SELECT * FROM health_conditions_under_18 WHERE AGE = '0-4 years' LIMIT 10"
asthma_data = pd.read_sql_query(query1, conn)

# Display the result as a DataFrame
asthma_data
```

```python
# Count the number of records for 'ADHD among persons under 18 years'
query2 = "SELECT COUNT(*) as ADHD_Record_Count FROM health_conditions_under_18 WHERE PANEL = 'ADHD among persons under 18 years'"
adhd_count = pd.read_sql_query(query2, conn)

# Display the result as a DataFrame
adhd_count
```

```python
# Define a custom sort order for the 'AGE' column
age_sort_order = {
    '0-4 years': 1,
    '5-9 years': 2,
    '5-17 years': 3,
    '10-17 years': 4,
    'Under 18 years': 5
}

# Query to group by 'AGE' and 'PANEL', calculate the average 'ESTIMATE' and count records
query3 = """
SELECT AGE, PANEL, COUNT(*) as Panel_Count
FROM health_conditions_under_18
GROUP BY AGE, PANEL
"""

# Execute the query
age_panel_grouped_data = pd.read_sql_query(query3, conn)

# Sort the DataFrame based on the custom order and then by 'PANEL'
age_panel_grouped_data['AGE_Sort_Order'] = age_panel_grouped_data['AGE'].map(age_sort_order)
sorted_data = age_panel_grouped_data.sort_values(by=['AGE_Sort_Order', 'PANEL'])

# Drop the helper column used for sorting
sorted_data = sorted_data.drop(columns=['AGE_Sort_Order'])

# Display the sorted result
sorted_data
```

```python
# Sort the records by 'PANEL_NUM' and return the top 5 records
query4 = "SELECT * FROM health_conditions_under_18 ORDER BY PANEL_NUM DESC LIMIT 5"
top_5_panel_num = pd.read_sql_query(query4, conn)

# Display the result as a DataFrame
top_5_panel_num
```

```python
# Close the database connection
conn.close()
```
